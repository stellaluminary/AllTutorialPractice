## 2. Color palette choices

URL : https://seaborn.pydata.org/examples/color_palettes.html

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style='white', context='talk')
rs = np.random.RandomState(8)

f, (ax1, ax2, ax3) = plt.subplots(3,1, figsize=(7,5), sharex=True)

x=np.array(list("ABCDEFGHIJ"))
y1=np.arange(1,11)
sns.barplot(x=x, y=y1, palette='rocket', ax=ax1)
ax1.axhline(0, color='k', clip_on=False)
ax1.set_ylabel('Sequential')

y2=y1-5.5
sns.barplot(x=x, y=y2, palette='vlag', ax=ax2)
ax2.axhline(0,color='k', clip_on=False)
ax2.set_ylabel('Diverging')

y3 = rs.choice(y1, len(y1), replace=False)
sns.barplot(x=x, y=y3, palette="deep", ax=ax3)
ax3.axhline(0, color="k", clip_on=False)
ax3.set_ylabel("Qualitative")

sns.despine(bottom=True)
plt.setp(f.axes, yticks=[])
plt.tight_layout(h_pad=2)
```

## 3. Different cubehelix palettes

URL : https://seaborn.pydata.org/examples/cubehelix_palette.html

```python
sns.set(style='dark')
rs = np.random.RandomState(50)

f, axes= plt.subplots(3,3, figsize=(9,9), sharex=True, sharey=True)

for ax,s in zip(axes.flat, np.linspace(0,3,10)):
    cmap = sns.cubehelix_palette(start=s, light=1, as_cmap=True)
    
    x,y=rs.randn(2,50)
    sns.kdeplot( x,y, cmap=cmap, shade=True, cut=5, ax=ax)
    ax.set(xlim=(-3,3), ylim=(-3,3))
    
f.tight_layout()
```

## 4. Scatterplot with categorical and numerical semantics

URL : https://seaborn.pydata.org/examples/different_scatter_variables.html

```python
sns.set(style="whitegrid")

diamonds = sns.load_dataset("diamonds")
diamonds.head()
```

```python
diamonds.shape
```

```python
diamonds.dtypes
```

```python
for i in [j for j in diamonds.columns if diamonds[j].dtypes == 'object']:
    print(i,'col unique:',diamonds[i].unique())
```

```python
f, ax = plt.subplots(figsize=(6.5, 6.5))
sns.despine(f, left=True, bottom=True)
clarity_ranking = ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]
sns.scatterplot(x="carat", y="price",
                hue="clarity", size="depth",
                palette="ch:r=-.2,d=.3_r",
                hue_order=clarity_ranking,
                sizes=(1,8), linewidth=0,
                data=diamonds, ax=ax)
                
```

```python
f, ax = plt.subplots(figsize=(6.5, 6.5))
sns.despine(f, left=True, bottom=True)
clarity_ranking = ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]
sns.scatterplot(x="carat", y="price",
                hue="clarity", size="depth",
                #palette="ch:r=-.2,d=.3_r",
                hue_order=clarity_ranking,
                sizes=(1,8),  linewidth=0,
                data=diamonds, ax=ax)
                
```

## 5. Distribution plot options

URL : https://seaborn.pydata.org/examples/distplot_options.html

```python
sns.set(style="white", palette="muted", color_codes=True)
rs = np.random.RandomState(10)
```

```python
f, axes=plt.subplots(2,2, figsize=(7,7), sharex=True)
sns.despine(left=True)

d=rs.normal(size=100)

sns.distplot(d, kde=False, color='b', ax=axes[0,0])
sns.distplot(d, hist=False, rug=True, color='r', ax=axes[0,1])
sns.distplot(d, hist=False, color='g', kde_kws={'shade':True}, ax=axes[1,0])
sns.distplot(d, color='m', ax=axes[1,1])
```

## 9. Hexbin plot with marginal distributions
URL : https://seaborn.pydata.org/examples/hexbin_marginals.html

```python
sns.set(style='ticks')

rs=np.random.RandomState(11)
x = rs.gamma(2, size=1000)
y=-.5*x + rs.normal(size=1000)
sns.jointplot(x,y,kind='hex', color="#4CB391")
#kind must be either 'scatter', 'reg', 'resid', 'kde', or 'hex'
```

```python
sns.set(style='ticks')

rs=np.random.RandomState(11)
x = rs.gamma(2, size=1000)
y=-.5*x + rs.normal(size=1000)

f, ax = plt.subplots(2,1)

sns.jointplot(x,y, kind='reg', ax=ax[0])
#kind must be either 'scatter', 'reg', 'resid', 'kde', or 'hex'
sns.jointplot(x,y, kind='kde', ax=ax[1])
```

## 10. Horizontal bar plots

URL : https://seaborn.pydata.org/examples/horizontal_barplot.html

```python
sns.set(style='whitegrid')

crashes = sns.load_dataset('car_crashes')
crashes.head()
```

```python
crashes = crashes.sort_values('total', ascending=False)
crashes.head()
```

```python
f,ax=plt.subplots(figsize=(6,15))

sns.set_color_codes('pastel')
sns.barplot(x='total', y='abbrev', data=crashes,
           label='Total', color='b')

sns.set_color_codes('muted')
sns.barplot(x='alcohol', y='abbrev', data=crashes,
           label='Alcohol-involved', color='b')

ax.legend(ncol=2, loc='lower right', frameon=True)
ax.set(xlim=(0,24), ylabel='',
      xlabel='Automobile collisions per billion miles')
sns.despine(left=True, bottom=True)
```
